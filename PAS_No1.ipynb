```python
import pandas as pd
from google.colab import drive #memanggil package import drive
drive.mount('/content/drive')
```

```python
path = '/content/drive/MyDrive/CSV/Life Expectancy Data3c - Sheet1.csv'


df = pd.read_csv(path)

df.describe()
```

```python
df
```

```python
#untuk menampilkan scatter
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Life expectancy', y='percentage expenditure')

plt.title('Grafik Scatter')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(65, 70)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.array(df['Life expectancy'])
y = np.array(df['percentage expenditure'])

print(f'x : {x}')
print(f'y : {y}')
```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
from sklearn.linear_model import LinearRegression #memanggil model regresi linear

model = LinearRegression() #variabel model bisa diganti
model.fit(x, y) 
```

```python
x_vis = np.array([0, 80]).reshape(-1, 1) #pembuatan garis merah dari x
y_vis = model.predict(x_vis) #model mesin lerning
```

```python
plt.scatter(x, y)
plt.plot(x_vis, y_vis, '-r')
plt.title('Grafik')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(60, 70)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```

```python
#untuk menampilkan scatter
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Population', y='GDP')

plt.title('Grafik Scatter')
plt.xlabel('Population')
plt.ylabel('GDP')
plt.xlim(0, 260000000)
plt.ylim(0, 10)
plt.grid(True)
plt.show()
```
